## Importing Libraries:

```python
import pandas
import tensorflow as tf
import cv2
import numpy as np
from keras.optimizers import Adam,RMSprop,SGD
from keras.models import load_model
from keras.optimizers import RMSprop
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, Input, MaxPool2D,ZeroPadding2D, BatchNormalization, Activation, MaxPooling2D, Flatten, Dense,Dropout
from keras.models import Model, load_model
from keras.callbacks import TensorBoard, ModelCheckpoint, EarlyStopping
```

### Image Data Augumentation for efficiency

```python
aug = ImageDataGenerator(rescale=1.5/255,
                                   rotation_range=25,
                                   width_shift_range=0.25,
                                   height_shift_range=0.35,
                                   horizontal_flip=True)
```

```python
train_data = aug.flow_from_directory("Dataset/train/", batch_size=10,target_size=(150,150))
```

```python
test_data = aug.flow_from_directory("Dataset/test/",batch_size=10,target_size=(150,150))
```

  ### VGGNet architecture

```python
model = Sequential()
model.add(Conv2D(input_shape=(150,150,3),filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(Conv2D(filters=64,kernel_size=(3,3),padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(Conv2D(filters=512, kernel_size=(3,3), padding="same", activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),strides=(2,2)))
model.add(Flatten())
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=4096,activation="relu"))
model.add(Dense(units=2, activation="softmax"))
```

### Using RMSProp Optimizer

```python
opt = RMSprop(learning_rate=0.1)
```

```python
model.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit_generator(train_data, epochs=10,shuffle = True,validation_data=test_data)
```

```python
# Saving this model to the directory
model.save("RMSprop.model")
```

Using RMSProp Optimizer we got a model accuracy of 60.77 % with 10 epochs.

## OpenCv using ROI for live detection

```python
#To access the saved model 
model=load_model("Adam.model")
```

### Preprocessing

```python
output={0:'without mask',1:'mask'}
col_dict={0:(0,0,255),1:(0,255,0)}
size = 4
```

```python
vid = cv2.VideoCapture(0) 
```

```python
# Cascade classifier using cv2 haarcasacade 
cas_clasifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
while True:
    (rval, im) = vid.read()
    im=cv2.flip(im,1,1) 
    new_size = cv2.resize(im, (im.shape[1] // 4, im.shape[0] // 4))
    
    faces = cas_clasifier.detectMultiScale(new_size)
    #can detect multiple faces
    for f in faces:
        (x, y, w, h) = [v * size for v in f] 
        
        #Extracting th face of the object
        face_img = im[y:y+h, x:x+w]
        rerect_sized=cv2.resize(face_img,(150,150))
        
        #normalization for accurate detection
        normalized=rerect_sized/255.0
        reshaped=np.reshape(normalized,(1,150,150,3))
        reshaped = np.vstack([reshaped])
        result=model.predict(reshaped)

        if result[0][0] < 0.25:
            pred = 1
        else:
            pred =0
        
        #print(result)   
      
        cv2.rectangle(im,(x,y),(x+w,y+h),col_dict[pred],2)
        cv2.rectangle(im,(x,y-40),(x+w,y),col_dict[pred],-1)
        cv2.putText(im, output[pred], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)

    cv2.imshow('',im)
    
    key = cv2.waitKey(10)
    
    if key == 27: 
        break

vid.release()

cv2.destroyAllWindows()
```
