```python
'''
Author: Asha Nagireddy
Code: Crime solving
version: 1
Date: 15 october 2019
'''
import math
import random

#function to calculate total time in seconds
def seconds(time):
    hours = int(time[:2])
    minutes = int(time[3:5])
    seconds = int(time[6:])
    total_sec = (3600*hours)+(60*minutes)+seconds
    return total_sec

#function to calculate distance
def distance(a,b):
    a = a[1:len(a)-1].split(",")
    b = b[1:len(b)-1].split(",")
    dist = abs(math.sqrt(((int(b[0]) - int(a[0]))**2) + (int(b[1]) - int(a[1]))**2))
    return dist

#extracting data from txt file
file = open("/Users/saiakhilparvathareddy/Desktop/evidence.txt", "r")
data = file.readlines()
file.close()
list1 =[]
list2 =[]
list3 = []
list4 =[]
list5 =[]

# storing data into 3 seperate lists
for i in data:
    i = i[1:].split(",")
    #removing ) from y coordinate
    j = i[1]
    j = j[:len(j)-1]
    i[1] = j
    #removing \n at the end of time value
    k = i[2]
    k = k[:len(k)-1]
    i[2] = k
    
    list1.append(i[0])
    list2.append(i[1])
    list3.append(i[2])

#taking user input of location and crime time
grid_location = input("Please enter the grid location in (x,y) format:")
crime_time = input("Please enter the crime time in hh:mm:ss format:")

#finding times before and after crime
list6 = list3 + [crime_time]
list6.sort()
index = list6.index(crime_time)
before_time = list6[index-1]
l1 = list3.index(before_time)
after_time = list6[index+1]
l2 = list3.index(after_time)

#calculating time in seconds
time_c = seconds(crime_time)
time_b = seconds(before_time)
time_a = seconds(after_time)

#calculating time difference
time1 = abs(time_c - time_b)
time2 = abs(time_c - time_a)

str1 = "(" + list1[l1]+","+list2[l1]+")"
str2 = "(" + list1[l2]+","+list2[l2]+")"

#calculating distance of 2 times from grid location
distance1 = distance(grid_location,str1)
distance2 = distance(grid_location,str2)


#finding speed of person
speed1 = distance1/time1
speed2 = distance2/time2


if (speed1 > 3) or (speed2 > 3):
    print("The person did not commit the crime")
elif (speed1 < 3) and (speed2 < 3):
    print("The person could have commited the crime")
```
